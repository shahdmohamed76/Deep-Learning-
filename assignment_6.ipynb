```python
import numpy as np
```

```python

# Input sentence (tokenized)
words = ["What", "are", "the", "symptoms", "of", "diabetes"]

# Step 1: Random embeddings for each word (for simplicity)
np.random.seed(0)
embeddings = np.random.rand(len(words), 4)  # 4-dimensional embeddings
```

```python
# Step 2: Initialize Weight matrices (Wq, Wk, Wv)
Wq = np.random.rand(4, 4)
Wk = np.random.rand(4, 4)
Wv = np.random.rand(4, 4)
```

```python
# Step 3: Compute Query, Key, Value
Q = np.dot(embeddings, Wq)
K = np.dot(embeddings, Wk)
V = np.dot(embeddings, Wv)
```

```python
# Step 4: Compute Attention Scores
scores = np.dot(Q, K.T)

# Step 5: Scale scores
d_k = Q.shape[1]
scores = scores / np.sqrt(d_k)
```

```python
# Step 6: Softmax function
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / exp_x.sum(axis=1, keepdims=True)

attention_weights = softmax(scores)

# Step 7: Final output (Self-Attention result)
output = np.dot(attention_weights, V)

# Print results
print("Attention Weights:\n", attention_weights)
print("\nOutput:\n", output)
```
