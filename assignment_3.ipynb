```python
import numpy as np
```

```python
# activation function 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
# input sequence 
X = [1, 2, 3]

#LSTM parameters

Wf = 0.5
Wi = 0.6
Wo = 0.8
Wc = 0.7

Uf = 0.1
Ui = 0.2
Uo = 0.4
Uc = 0.3

bf = 0
bi = 0
bo = 0
bc = 0

# initial states
h = 0
c = 0
```

```python
# 
for t, x in enumerate(X):

    # Forget gate:
    f = sigmoid(Wf * x + Uf * h + bf)

    # Input gate:
    i = sigmoid(Wi * x + Ui * h + bi)

    # Candidate Cell State:
    c_tilde = tanh(Wc * x + Uc * h + bc)

    # Cell State Update:
    c = f * c + i * c_tilde

    # Output Gate:
    o = sigmoid(Wo * x + Uo * h + bo)

    # Hidden State Update:
    h = o * tanh(c)

    print(f"Time step {t+1}")
    print("Forget gate:", f)
    print("Input gate:", i)
    print("Candidate:", c_tilde)
    print("Cell state:", c)
    print("Output gate:", o)
    print("Hidden state:", h)
    print()
```

```python
# Predict the Next Value
Wy = 4
by = 0

y_pred = Wy * h + by

print("Predicted next value:", y_pred)
```
